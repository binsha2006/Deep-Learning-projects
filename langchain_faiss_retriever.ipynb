```python
!pip install langchain langchain-community sentence-transformers faiss-cpu
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings
sentences = [
    "Virat Kohli scored a century in the cricket match.",
    "The bowler took five wickets in the tournament.",
    "Cricket is one of the most popular sports in India.",
    "The team celebrated their victory after the final match.",
    "Cooking pasta requires boiling water first.",
    "Baking cakes needs proper temperature control.",
    "Fresh vegetables improve the taste of any recipe.",
    "Python is widely used for machine learning.",
    "Functions help organize code into reusable blocks.",
    "Debugging is an important programming skill."
]
embedding_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
db = FAISS.from_texts(sentences, embedding_model)
retriever = db.as_retriever(search_kwargs={"k": 3})
queries = [
    "best cricket player",
    "how to bake a cake",
    "python coding functions"
]
for query in queries:
    print("\n" + "=" * 60)
    print("QUERY:", query)
    print("=" * 60)
    results = retriever.invoke(query)
    for doc in results:
        print(doc.page_content)
        print()
print("Manual FAISS and LangChain FAISS produce similar semantic results.")
```
