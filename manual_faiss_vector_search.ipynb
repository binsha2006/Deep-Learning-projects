```python
!pip install faiss-cpu sentence-transformers
import faiss
import numpy as np
import pickle
from sentence_transformers import SentenceTransformer
sentences = [
    "Virat Kohli scored a century in the cricket match.",
    "The bowler took five wickets in the tournament.",
    "Cricket is one of the most popular sports in India.",
    "The team celebrated their victory after the final match.",
    "Cooking pasta requires boiling water first.",
    "Baking cakes needs proper temperature control.",
    "Fresh vegetables improve the taste of any recipe.",
    "Python is widely used for machine learning.",
    "Functions help organize code into reusable blocks.",
    "Debugging is an important programming skill."
]
model = SentenceTransformer("all-MiniLM-L6-v2")
embeddings = model.encode(sentences)
dimension = embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(np.array(embeddings).astype("float32"))
queries = [
    "best cricket player",
    "how to bake a cake",
    "python coding functions"
]
for query in queries:
    print("\n" + "=" * 60)
    print("QUERY:", query)
    print("=" * 60)
    query_embedding = model.encode([query])
    distances, indices = index.search(
        np.array(query_embedding).astype("float32"),
        k=2
    )
    for i in range(2):
        print("Sentence:", sentences[indices[0][i]])
        print("Distance:", distances[0][i])
        print()
faiss.write_index(index, "faiss_index.bin")
with open("documents.pkl", "wb") as file:
    pickle.dump(sentences, file)
loaded_index = faiss.read_index("faiss_index.bin")
with open("documents.pkl", "rb") as file:
    loaded_sentences = pickle.load(file)
print("\nINDEX RELOADED SUCCESSFULLY")
```
