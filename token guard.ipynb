```python
import tiktoken

class ChatBot:
    def __init__(self):
        self.history = []
        self.encoding = tiktoken.get_encoding("cl100k_base")
    def count_tokens(self):
        """Count total tokens in chat history"""
        total_text = ""
        for msg in self.history:
            total_text += f"{msg['role']}: {msg['content']}\n"
        tokens = self.encoding.encode(total_text)
        return len(tokens)
    def trim_history(self):
        """
        Keep only last 4 exchanges (8 messages)
        if token count exceeds 3000
        """
        total_tokens = self.count_tokens()
        if total_tokens > 3000:
            self.history = self.history[-8:]
            print("\n⚠️ History exceeded 3000 tokens.")
            print("⚠️ Older messages removed. Keeping last 4 exchanges only.\n")
    def chat(self, user_input, bot_response):
        self.history.append({
            "role": "user",
            "content": user_input
        })
        self.history.append({
            "role": "assistant",
            "content": bot_response
        })
        print(f"Bot: {bot_response}")
        total_tokens = self.count_tokens()
        print(f"\nTotal Tokens: {total_tokens}")
        self.trim_history()
bot = ChatBot()
for i in range(20):
    user_msg = f"User message number {i} " * 40
    bot_msg = f"Bot response number {i} " * 40
    bot.chat(user_msg, bot_msg)
print("\nFinal History Length:", len(bot.history))
```
