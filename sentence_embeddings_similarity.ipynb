```python
!pip install sentence-transformers scikit-learn
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
sentences = [

    "Virat Kohli scored a century in the cricket match.",
    "The bowler took five wickets in the tournament.",
    "Cricket is one of the most popular sports in India.",
    "The team celebrated their victory after the final match.",
    "Cooking pasta requires boiling water first.",
    "Baking cakes needs proper temperature control.",
    "Fresh vegetables improve the taste of any recipe.",
    "Python is widely used for machine learning.",
    "Functions help organize code into reusable blocks.",
    "Debugging is an important programming skill."
]
model = SentenceTransformer("all-MiniLM-L6-v2")
embeddings = model.encode(sentences)
similarity_matrix = cosine_similarity(embeddings)
df = pd.DataFrame(
    similarity_matrix,
    index=[f"S{i+1}" for i in range(len(sentences))],
    columns=[f"S{i+1}" for i in range(len(sentences))]
)
print(df)
```
