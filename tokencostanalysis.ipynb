```python
!pip install tiktoken
import tiktoken
import pandas as pd
encoding = tiktoken.get_encoding("cl100k_base")
def token_cost(text, price_per_million=0.10):
    tokens = encoding.encode(text)
    token_count = len(tokens)
    cost = (token_count / 1_000_000) * price_per_million
    return token_count, cost
paragraph = """
Artificial intelligence is transforming industries across the world by enabling
machines to perform tasks that normally require human intelligence. Machine learning,
a subset of AI, focuses on teaching computers to learn patterns from data and improve
their performance over time without being explicitly programmed. Deep learning models,
which are inspired by the structure of the human brain, are capable of processing
large amounts of complex information such as images, text, and speech. AI is widely
used in healthcare for disease prediction, in finance for fraud detection, in retail
for recommendation systems, and in transportation for self-driving technology.

As organizations continue to adopt AI solutions, the demand for skilled professionals
in data science, machine learning, and artificial intelligence is growing rapidly.
Python has become one of the most popular programming languages for AI development
because of its simplicity and powerful libraries such as NumPy, pandas, TensorFlow,
and PyTorch. Data preprocessing, feature engineering, model training, and evaluation
are important stages in building successful machine learning systems. Engineers also
need to understand concepts like overfitting, underfitting, bias, variance, and
hyperparameter tuning to improve model performance.

Despite its advantages, AI also raises ethical concerns including data privacy,
algorithmic bias, transparency, and job displacement. Researchers and governments are
working together to establish regulations and responsible AI practices to ensure that
technology benefits society fairly and safely.
"""
python_script = "\n".join([
    f"print('This is line {i}')"
    for i in range(1, 51)
])
conversation = """
User: Hello, how are you?
Assistant: I am doing well. How can I help you today?
User: Can you explain machine learning?
Assistant: Machine learning is a branch of AI where computers learn patterns from data.
User: Give me an example.
Assistant: Spam email detection is a common machine learning application.
"""
results = []
test_cases = {
    "300-word Paragraph": paragraph,
    "50-line Python Script": python_script,
    "3-turn Conversation": conversation
}
for name, text in test_cases.items():
    tokens, cost = token_cost(text)
    results.append({
        "Input Type": name,
        "Token Count": tokens,
        "Estimated Cost ($)": round(cost, 8)
    })
df = pd.DataFrame(results)
print("\nTOKEN COST COMPARISON TABLE\n")
print(df.to_string(index=False))
```
