```python
!pip install tiktoken
import tiktoken
encoding = tiktoken.get_encoding("cl100k_base")
MODEL_LIMIT = 8192
def estimate_context_usage(messages_list):
    total_tokens = 0
    for message in messages_list:
        content = message["content"]
        tokens = encoding.encode(content)
        total_tokens += len(tokens)
    usage_percent = (total_tokens / MODEL_LIMIT) * 100
    print("=" * 60)
    print("CONTEXT WINDOW USAGE")
    print("=" * 60)
    print(f"Total Tokens Used: {total_tokens}")
    print(f"Model Limit: {MODEL_LIMIT}")
    print(f"Usage Percentage: {usage_percent:.2f}%")
    if usage_percent > 80:
        print("\nWARNING: Context window usage exceeds 80%!")
        print("The model may soon run out of memory space.")
    else:
        print("\nContext usage is within safe limits.")
messages = []
for i in range(40):
    messages.append({
        "role": "user",
        "content": (
            "Machine learning is a branch of artificial intelligence that "
            "allows systems to learn from data and improve over time. "
        ) * 10
    })
estimate_context_usage(messages)
```
