```python
!pip install tiktoken
import tiktoken
encoding = tiktoken.get_encoding("cl100k_base")
inputs = {
    "English Sentence": "Machine learning models are becoming more powerful every year.",
    "Python Function": """
def add_numbers(a, b):
    return a + b
""",
    "Native Language Sentence": "എനിക്ക് മെഷീൻ ലേണിംഗ് പഠിക്കാൻ ഇഷ്ടമാണ്",
    "Number": "1234567",
    "Email Address": "student123@gmail.com",
    "Mathematical Notation": "f(x) = x^2 + 3x - 5"
}
for name, text in inputs.items():
    tokens = encoding.encode(text)
    token_strings = [encoding.decode([token]) for token in tokens]
    print("=" * 60)
    print(f"{name}")
    print("-" * 60)
    print("Input:")
    print(text)
    print("\nTokens:")
    print(token_strings)
    print("\nToken Count:", len(tokens))
    print("=" * 60)
    print()
```
