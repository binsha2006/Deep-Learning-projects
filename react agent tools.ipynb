```python
from groq import Groq
client = Groq(
    api_key="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"
)
facts = {
    "population of india": "1.44 billion",
    "speed of light": "3x10^8 m/s"
}
def calculate(expression):
    return eval(expression)
def lookup(fact):
    fact = fact.lower()
    if fact in facts:
        return facts[fact]
    return "Fact not found"
question = """
What is double the population of India in billions?
"""
messages = [
    {
        "role": "system",
        "content": """
You are a ReAct AI agent.
You can use two actions:
1. lookup(fact)
2. calculate(expression)
Follow this exact format:
Thought:
Action:
Observation:
Repeat until final answer.
Available facts:
- population of india
- speed of light
"""
    },
    {
        "role": "user",
        "content": question
    }
]
response = client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=messages,
    temperature=0.3,
    max_tokens=300
)
agent_output = response.choices[0].message.content
print("=" * 70)
print("INITIAL AGENT OUTPUT")
print("=" * 70)
print(agent_output)
print("\n")
fact_result = lookup("population of india")
print("=" * 70)
print("LOOKUP ACTION")
print("=" * 70)
print("lookup('population of india')")
print("Observation:", fact_result)
print("\n")
calculation_result = calculate("1.44 * 2")
print("=" * 70)
print("CALCULATION ACTION")
print("=" * 70)
print("calculate('1.44 * 2')")
print("Observation:", calculation_result)
print("\n")
print("=" * 70)
print("FINAL ANSWER")
print("=" * 70)
print(f"Double the population of India is approximately {calculation_result} billion.")
```
