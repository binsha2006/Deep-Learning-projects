```python
import tiktoken
class ChatBot:
    def __init__(self):
        self.history = []
        self.last_retrieved_chunks = []
        self.encoding = tiktoken.get_encoding("cl100k_base")
    def count_tokens(self):
        total_text = ""
        for msg in self.history:
            total_text += f"{msg['role']}: {msg['content']}\n"
        tokens = self.encoding.encode(total_text)
        return len(tokens)
    def trim_history(self):
        total_tokens = self.count_tokens()

        if total_tokens > 3000:
            self.history = self.history[-8:]
            print("\n⚠️ History exceeded 3000 tokens.")
            print("⚠️ Older messages removed. Keeping last 4 exchanges only.\n")
    def summarize_conversation(self):
        print("\n Conversation Summary:")
        print("- Discussed chatbot conversation flow")
        print("- Implemented token counting using tiktoken")
        print("- Added automatic memory trimming feature\n")
    def show_sources(self):
        print("\n Last Retrieved Chunks:")
        if not self.last_retrieved_chunks:
            print("No chunks retrieved yet.\n")
            return
        for i, chunk in enumerate(self.last_retrieved_chunks, start=1):
            print(f"{i}. {chunk}")
        print()
    def chat(self, user_input):
        if user_input.lower() == "summary":
            self.summarize_conversation()
            return
        if user_input.lower() == "sources":
            self.show_sources()
            return
        self.history.append({
            "role": "user",
            "content": user_input
        })
        self.last_retrieved_chunks = [
            "Chunk about tokenization and embeddings.",
            "Chunk explaining FAISS vector database.",
            "Chunk related to chatbot memory management."
        ]
        bot_response = f"Bot response to: {user_input}"
        self.history.append({
            "role": "assistant",
            "content": bot_response
        })
        print(f"\nBot: {bot_response}")
        total_tokens = self.count_tokens()
        print(f"Total Tokens: {total_tokens}")
        self.trim_history()
bot = ChatBot()
bot.chat("What is tokenization?")
bot.chat("Explain embeddings")
bot.chat("How does FAISS work?")
bot.chat("summary")
bot.chat("sources")
bot.chat("Explain memory trimming in chatbots")
```
